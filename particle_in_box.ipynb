### Defining potentials for schordinger equation

```python
# defining constants
h = 1
m = 1
a = 1
V = 0
psi = 0
dpsi = 1 # first derivative of psi
x = 0
dx = a*0.001
xlist = []
psilist = []
E = 4.9*3**2
while x< a:
    ddpsi = 2*m/h**2*(V-E)*psi   #second derivative of psi
    dpsi = dpsi + ddpsi*dx
    psi = psi + dpsi*dx
    x = x + dx
    xlist.append(x)
    psilist.append(psi)
    plt.plot(xlist,psilist)

```

```python
# defining constants
h = 1
m = 1
a = 1
V = 0


dx = a*0.001
E = 0
dE = 0.01
counter = 1
nmax = 5
eigenfunctions = []
eigenenergies = []
eigenfunctionsxlist = []
while counter <=nmax:
    psi = 1
    while abs(psi) > 0.001:
        psi = 0
        dpsi = 1 # first derivative of psi
        x = 0
        E = E + dE
        xlist = []
        psilist = []
        while x< a:
            ddpsi = 2*m/h**2*(V-E)*psi   #second derivative of psi
            dpsi = dpsi + ddpsi*dx
            psi = psi + dpsi*dx
            x = x + dx
            xlist.append(x)
            psilist.append(psi)
    counter += 1
    eigenfunctions.append([psilist])
    eigenfunctionsxlist.append([xlist])
    eigenenergies.append(E)
    E = E*1.1
```

# plotting eigenenergies 

```python
# defining constants
h = 1
m = 1
a = 1
V = 0

for E in eigenenergies:
    psi = 0
    dpsi = 1 # first derivative of psi
    x = 0
    dx = a*0.001
    xlist = []
    psilist = []
    while x< a:
        ddpsi = 2*m/h**2*(V-E)*psi   #second derivative of psi
        dpsi = dpsi + ddpsi*dx
        psi = psi + dpsi*dx
        x = x + dx
        xlist.append(x)
        psilist.append(psi)
        plt.plot(xlist,psilist)
```
